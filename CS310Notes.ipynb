##Tree Types
---

Balanced Binary Search Tree- BST Property & Height differ at most by 1

Perfect Binary Tree -> Complete/Full/ Nothing missing

Complete binary tree -(leaf Nodes differ by at most 1 && from Left --> Right at lowest level)

Full binary tree -> 0/2 Nodes


---


Full Binary Tree: Every node has either 0 or 2 child nodes, i.e., left and right or no children

Complete Binary Tree: All levels, except possibly the last, are filled, and all nodes are as left as possible

Perfect Binary Tree: All nodes have exactly two children, and all leaf nodes are at the same level

Balanced Binary Tree: The heights of any node's left and right subtrees differ by at most one

Binary Search Tree (BST): A binary tree in which the left child node is smaller, and the right child node is greater than the parent node

Threaded Binary Tree: Each node of the binary tree contains extra information, called threads, that allows for efficient tree traversal without recursion or a stack

Heap: A complete binary tree that satisfies the heap property

## Heap Sort
Procedure
1.   Organize into Heap O(n)
2.   Transfer into Ascending Order O(nlogn)

Organize into Heap
* Runtime nlog(n)
* Stack require extra space
* Space Complexity Nice

Heapify/Build Heap
* Turning each subtree into a heap
 * Case 1: Leaf
   * nodes are fine
 * Case 2: Internal Node
   * Percalate Down

Time Complexity: O(n)
* B/c tree is complete we can skip N/2 elements
* O(n)
 * n/2 does require 0 percalate
 * n/4 require at most 1 per.
 * n/8 require at most 2 per.

Proof: https://stackoverflow.com/questions/9755721/how-can-building-a-heap-be-on-time-complexity

Note: Percalate Up from leaf will not work b/c the the top is not a heap

Note: Starting from the top and percalate up is a possible option but requires percalating on leaf nodes

Note: Array-Implementation just percalate down starting from the end of array


##Phase 2 (Turning into ascending order)
* Continue extracting the max value from the heap and placeing to the end
* In place sorting algorithm (Space Complexity O(1) )

Note: When transferring to ascending order, size of array always bc 1 smaller. That one additional space can be used to store the next biggest element

````python
%%writefile heapsort.cpp

Heapsort(numbers, numbersSize) {
   // Heapify numbers array
   for (i = numbersSize / 2 - 1; i >= 0; i--) {
      MaxHeapPercolateDown(i, numbers, numbersSize)
   }
  //Turn into Ascending Order
   for (i = numbersSize - 1; i > 0; i--) {
      Swap numbers[0] and numbers[i]
      MaxHeapPercolateDown(0, numbers, i)
   }
}
````
